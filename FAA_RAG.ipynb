```python
from langchain_community.document_loaders import PyPDFLoader
import os
```

```python
def load_pdf(path,category):
  loader = PyPDFLoader(path)
  docs = loader.load()
  for d in docs:
    d.metadata["source"]=os.path.basename(path)
    d.metadata["category"]=category
  return docs

documents=[]
documents+=load_pdf("/content/drive/MyDrive/data/aerospacedata/FAA_Installation,inspection.pdf","Inspection")
documents+=load_pdf("/content/drive/MyDrive/data/aerospacedata/FAA_prevention,inspection.pdf","Inspection")
documents+=load_pdf("/content/drive/MyDrive/data/aerospacedata/FAA_pervention,inspection(1).pdf","Inspection")
documents+=load_pdf("/content/drive/MyDrive/data/aerospacedata/FAA_prevention,inpection(2).pdf","Inspection")

documents+=load_pdf("/content/drive/MyDrive/data/aerospacedata/FAA_composite structures.pdf","Structures")
documents+=load_pdf("/content/drive/MyDrive/data/aerospacedata/FAA_metallic structures.pdf","Structures")

documents+=load_pdf("/content/drive/MyDrive/data/aerospacedata/FAA_Alterations and Repairs.pdf","Repairs")
documents+=load_pdf("/content/drive/MyDrive/data/aerospacedata/FAA_Repair station training.pdf","Repairs")

documents+=load_pdf("/content/drive/MyDrive/data/aerospacedata/FAA_Damage Tolerance.pdf","Damage")

documents+=load_pdf("/content/drive/MyDrive/data/aerospacedata/FAA_Thermal fatigue.pdf","Fatigue")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=400,
    chunk_overlap=80
)

chunks = splitter.split_documents(documents)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vectorstore = FAISS.from_documents(chunks, embeddings)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_classic.chains import create_retrieval_chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from transformers import BitsAndBytesConfig
from langchain_huggingface.llms import HuggingFacePipeline

model_id = "mistralai/Mistral-7B-Instruct-v0.1"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4"
)

tokenizer = AutoTokenizer.from_pretrained(model_id)

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="auto"
)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=512,
    temperature=0.1,
)

llm = HuggingFacePipeline(pipeline=pipe)
```

```python
prompt = ChatPromptTemplate.from_template("""
### ROLE
You are a highly precise Aerospace Engineering Assistant. Your goal is to provide technical answers based strictly on the provided documentation.

### CONSTRAINTS
1. ONLY use the provided context below.
2. If the context does not contain the answer, state: "I could not find this information in the provided documents."
3. Do not use outside knowledge or general industry assumptions.
4. Maintain technical terminology as it appears in the text.
5. Do not add any additional information.
6. Do not use prior knowledge.

### CONTEXT
{context}

### INSTRUCTIONS
- Answer the user's question clearly and concisely.
- If the context contains specific values, units (e.g., kN, PSI, Mach), or part numbers, include them exactly.
- If the context is insufficient but related, do not speculate.

Question: {input}

Answer:
""")
```

```python
document_chain = create_stuff_documents_chain(llm, prompt)
```

```python
query = "What are acceptable methods for aircraft crack inspection?"
```

```python
retriever = vectorstore.as_retriever(search_kwargs={"k": 5})
docs = retriever.invoke(query)

arr = []
for d in docs:
    print(d.metadata["source"])
    print(d.metadata["category"])
    arr.append(d.metadata["category"])

category_1=arr[0]
print(arr)
arr.clear()
```

```python
retriever = vectorstore.as_retriever(
    search_type="similarity_score_threshold",
    search_kwargs={
        "score_threshold": 0.5,
        "k": 5,
        "filter": {"category": category_1}
    }
)

docs = retriever.invoke(query)

rag_chain = create_retrieval_chain(retriever, document_chain)

if len(docs) == 0:
    print("No relevant documents found.")
else:
    response = rag_chain.invoke({"input": query})
    print(response["answer"])
```
